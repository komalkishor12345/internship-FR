```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager


service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)


driver.get('https://www.shine.com/')


job_title_field = driver.find_element(By.ID, 'id_q')
job_title_field.send_keys('Data Analyst')

location_field = driver.find_element(By.ID, 'id_loc')
location_field.send_keys('Bangalore')


search_button = driver.find_element(By.XPATH, '//button[contains(text(), "Search")]')
search_button.click()


time.sleep(5)


jobs = driver.find_elements(By.CLASS_NAME, 'jobTuple')[:10]

job_data = []
for job in jobs:
    job_title = job.find_element(By.CLASS_NAME, 'title').text
    job_location = job.find_element(By.CLASS_NAME, 'loc').text
    company_name = job.find_element(By.CLASS_NAME, 'company').text
    experience_required = job.find_element(By.CLASS_NAME, 'exp').text
    
    job_data.append({
        'Job Title': job_title,
        'Job Location': job_location,
        'Company Name': company_name,
        'Experience Required': experience_required
    })


df = pd.DataFrame(job_data)


print(df)


driver.quit()
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Setup the WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Step 1: Get the webpage
driver.get('https://www.naukri.com/')

# Step 2: Enter job title and location
job_title_field = driver.find_element(By.XPATH, '//input[@placeholder="Enter skills / designations / companies"]')
job_title_field.send_keys('Data Scientist')

location_field = driver.find_element(By.XPATH, '//input[@placeholder="Enter location"]')
location_field.send_keys('Bangalore')

# Step 3: Click the search button
search_button = driver.find_element(By.XPATH, '//button[contains(text(), "Search")]')
search_button.click()


time.sleep(5)


jobs = driver.find_elements(By.XPATH, '//article[@class="jobTuple bgWhite br4 mb-8"]')[:10]

job_data = []
for job in jobs:
    job_title = job.find_element(By.XPATH, './/a[@class="title fw500 ellipsis"]').text
    job_location = job.find_element(By.XPATH, './/li[@class="fleft grey-text br2 placeHolderLi location"]').text
    company_name = job.find_element(By.XPATH, './/a[@class="subTitle ellipsis fleft"]').text
    experience_required = job.find_element(By.XPATH, './/li[@class="fleft grey-text br2 placeHolderLi experience"]').text
    
    job_data.append({
        'Job Title': job_title,
        'Job Location': job_location,
        'Company Name': company_name,
        'Experience Required': experience_required
    })


df = pd.DataFrame(job_data)


print(df)


driver.quit()
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager


service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)


driver.get('https://www.flipkart.com/')
time.sleep(3)


try:
    close_button = driver.find_element(By.XPATH, '//button[contains(text(), "âœ•")]')
    close_button.click()
except:
    pass


search_field = driver.find_element(By.NAME, 'q')
search_field.send_keys('sunglasses')
search_icon = driver.find_element(By.CLASS_NAME, 'L0Z3Pu')
search_icon.click()
time.sleep(3)


sunglasses_data = []


def scrape_current_page():
    products = driver.find_elements(By.XPATH, '//div[contains(@class, "_1AtVbE")]//div[contains(@class, "_13oc-S")]')
    for product in products:
        try:
            brand = product.find_element(By.CLASS_NAME, '_2WkVRV').text
        except:
            brand = 'N/A'
        
        try:
            product_description = product.find_element(By.CLASS_NAME, 'IRpwTa').text
        except:
            product_description = 'N/A'
        
        try:
            price = product.find_element(By.CLASS_NAME, '_30jeq3').text
        except:
            price = 'N/A'
        
        sunglasses_data.append({
            'Brand': brand,
            'Product Description': product_description,
            'Price': price
        })

 
while len(sunglasses_data) < 100:
    scrape_current_page()
    try:
        next_button = driver.find_element(By.XPATH, '//a[contains(text(), "Next")]')
        next_button.click()
        time.sleep(3)
    except:
        print("No more pages found or some issue in navigating to the next page.")
        break

      
df = pd.DataFrame(sunglasses_data[:100]) 
print(df)


df.to_csv('sunglasses_data.csv', index=False)


driver.quit()
```
